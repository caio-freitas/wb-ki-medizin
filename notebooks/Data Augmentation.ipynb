```python
import pandas as pd
import numpy as np
from utils import load_ecg, get_target
from scipy.io import savemat
import matplotlib.pyplot as plt
import time
```

```python
def add_noise(signal: pd.Series, std: float):
    noise = np.random.normal(0, std, signal.shape)
    return signal+noise
    
```

```python
CURR_FILE = "training/train_ecg_02610.mat"
ecg_signal = load_ecg(CURR_FILE)
target = get_target("train_ecg_02610")
```

```python
plt.figure(1, (16,8))
plt.title(target)
plt.plot(ecg_signal)
plt.grid(which="major")
```

```python
noisy_signal = add_noise(ecg_signal, 10)
plt.figure(1, (16,8))
plt.title(target)
plt.plot(noisy_signal)
plt.grid(which="major")
```

```python
savemat(f"{time.time()}.mat", {"val": [noisy_signal]})
```
